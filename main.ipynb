```python
import csv
import pandas as pd
import tui
import matplotlib.pyplot as plt
import mypackage
import os
from mypackage import retrieve_data_by_id
from mypackage import get_listings_by_location
from mypackage import get_listings_by_property_type
from mypackage import show_listings_by_location
from mypackage import get_top_amenities
from mypackage import get_top_locations
from mypackage import get_top_locations_by_review_score
from mypackage import calculate_mean_min_nights_by_location
from mypackage import generate_bedroom_pie_chart
from mypackage import plot_room_type_counts
from mypackage import generate_accommodates_price_scatter
from mypackage import plot_airbnb_prices_by_year
from mypackage import plot_average_price_by_property_type

def load_data(file_path):
    # Function to retrieve CSV data
    try:
        with open(file_path) as csv_file:
            csv_data = list(csv.DictReader(csv_file))
            return csv_data
    except FileNotFoundError:
        print("File not found!")
    except Exception as e:
        print(str(e))
    
def run_airbnb_data_analysis():
    file_path = input("Enter the file path of the CSV file: ")
    
    while True:
        tui.display_main_menu()
        choice = input("\033[40mEnter your choice (1/2/3/4): \033[0m")
        if choice == '1':
            while True:
                tui.display_csv_module_options()
                csv_choice = input("\033[40mEnter your choice (1/2/3/4/5/6): \033[0m")
                if csv_choice == '1':
                    retrieve_data_by_id(file_path)

                elif csv_choice == '2':
                    location = input("Enter location: ")
                    if location == 'quit':
                        break
                    listings = get_listings_by_location(location, file_path)
                    for listing in listings:
                        print(f"Host Name: {listing['host_name']}\nProperty Type: {listing['property_type']}\nPrice: {listing['price']}\nMinimum Nights: {listing['minimum_nights']}\nMaximum Nights: {listing['maximum_nights']}\n")
             
                elif csv_choice == '3':
                    property_type = input("\033[40mEnter the property type: \033[0m")
                    listings = get_listings_by_property_type(property_type, file_path)
                    if listings:
                        for listing in listings:
                            print(f"Room Type: {listing['room_type']}\nAccommodates: {listing['accommodates']}\nBathrooms: {listing['bathrooms_text']}\nBedrooms: {listing['bedrooms']}\nBeds: {listing['beds']}\n")
                    else:
                        print(f"No listings found for property type '{property_type}'")
                    
                elif csv_choice == '4':
                    location = input("Enter location: ")
                    listings = show_listings_by_location(location, file_path)

                    if listings:
                        for listing in listings:
                            print(f"Host ID: {listing['Host ID']}\nLocation: {listing['Location']}\nLast Review: {listing['Last Review']}\nFirst Review: {listing['First Review']}\nHost Response Time: {listing['Host Response Time']}\nAccommodates: {listing['Accommodates']}\n")
                    else:
                        print("No listings found for this location.")
                    
                elif csv_choice == '5':
                    break
                else:
                    print("\033[40mInvalid choice! Please try again.\033[0m")

        elif choice == '2':
            while True:
                tui.display_pandas_module_options()
                pandas_choice = input("\033[40mEnter your choice (1/2/3/4/5): \033[0m")
                if pandas_choice == '1':
                    top_amenities = get_top_amenities(file_path)
                    print(top_amenities)
                elif pandas_choice == '4':
                    mean_min_nights = calculate_mean_min_nights_by_location(file_path)
                    print(mean_min_nights)

                elif pandas_choice == '3':
                    top_locations = get_top_locations_by_review_score(file_path)
                    print(top_locations)
                elif pandas_choice == '2':
                    top_locations = get_top_locations(file_path)
                    print(top_locations)

                elif pandas_choice == '5':
                    break
                else:
                    print("\033[40mInvalid choice! Please try again.\033[0m")

        elif choice == '3':
            while True:
                tui.display_visualization_module_options()
                visualization_choice = input("\033[40mEnter your choice (1/2/3/4/5/6): \033[0m")
                if visualization_choice == '1':
                    fig = generate_bedroom_pie_chart(file_path)
                    print("Pie chart generated and saved as 'bedroom_pie_chart.png'.")
                elif visualization_choice == '2':
                    fig = plot_room_type_counts()
                    print("Plot Room Type Counts:")
                elif visualization_choice == '3':
                    fig = generate_accommodates_price_scatter(file_path)
                    print("Scatter Plot generated and saved as 'accommodates_price_scatter.png'.")
                elif visualization_choice == '4':
                    fig = plot_airbnb_prices_by_year()
                    print("Plot Airbnb Prices by Year:")
                elif visualization_choice == '5':
                    fig = plot_average_price_by_property_type(file_path)
                    print("Plot the Average price per night for each property type:")
                elif visualization_choice == '6':
                    break
                else:
                    print("\033[40mInvalid choice! Please try again.\033[0m")
        elif choice == '4':
                break
        else:
            print("\033[40mInvalid choice! Please try again.\033[0m")


# Run the Airbnb Data Analysis Software
if __name__ == "__main__":
    run_airbnb_data_analysis()
```
